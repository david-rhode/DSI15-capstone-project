**BUILD A FINAL DATAFRAME OF FICTIONAL PROPERTIES COVERING ALL LONDON POSTCODES**

- We have found the most successful model based on the Land Registry transactions (which cover over 60k London postcodes). 

- Now we can construct a new dataframe for all 200k London postcodes, and build out features for each of them. 

- Multiply this x 4 property types (flats, semi-detached, terraced, detached) to create a dataframe of about 800k fictional properties. 

- We can then use our most successful model to predict prices for each of them, giving us a final matrix that will allow us to 'move' a property from one postcode to another.

- Write a 'convert_postcode' function to carry out this move and visualise it.


```python
import pandas as pd
import geopandas as gpd
import joblib
import matplotlib.pyplot as plt
```

```python
# start building the dataframe with a list of all unique London postcodes, turn them into Cartesian map coordinates

london_postcodes = pd.read_csv('../../../big_files/london_postcodes.csv')
london_postcodes = london_postcodes.drop(columns = ['Unnamed: 0'])
final_df = london_postcodes.copy()
final_df = gpd.GeoDataFrame(final_df, geometry = gpd.points_from_xy(final_df['longitude'], final_df['latitude']))
final_df = final_df.set_crs(epsg = 4326)
final_df = final_df.to_crs(epsg = 27700)
```

```python
# import shape file for all LSOA boundaries and set CRS

lsoa = gpd.read_file('../../../big_files/LSOA_2011_London_gen_MHW.shp')
lsoa = lsoa[['LSOA11CD', 'geometry']]
lsoa = lsoa.set_crs(epsg = 27700)
```

```python
# spatial join to merge LSOA shape file with postcodes

final_df = gpd.sjoin(final_df, lsoa, how = 'left', op = 'intersects')
final_df = final_df.rename(columns = {'LSOA11CD': 'lsoa'})
final_df = final_df.dropna()
final_df= final_df.drop(columns = ['index_right'])
```

```python
# import free school meal data and merge it with final dataframe

school_meals = pd.read_csv('../../../big_files/school_meals.csv')
school_meals = school_meals.rename(columns = {'pupil_perc_fsm_lsoa': 'fsm_lsoa'})
school_meals = school_meals.drop(columns = ['Unnamed: 0'])
final_df = pd.merge(final_df, school_meals, how = 'left', left_on = 'lsoa', right_on = 'lsoa')
```

```python
# import ward data to link to lsoa in final dataframe

ward_lsoa = pd.read_csv('../../../big_files/ward_lsoa.csv')
ward_lsoa = ward_lsoa.drop(columns = ['Unnamed: 0'])
final_df = pd.merge(final_df, ward_lsoa, how = 'left', left_on = 'lsoa', right_on = 'lsoa')
```

```python
# import crime survey data, set CRS, merge with final dataframe

bcp = gpd.read_file('../../../big_files/borough_crime_perception_gdf.shp')
bcp = bcp.drop(columns = ['GSS_CODE', 'HECTARES', 'NONLD_AREA', 'ONS_INNER', 'SUB_2009', 'SUB_2006'])
bcp = bcp.set_crs(epsg = 27700)
bcp = bcp.drop(columns = ['geometry'])
final_df = pd.merge(final_df, bcp, how = 'left', left_on = 'borough', right_on = 'borough_na')
```

```python
#data missing from City of London area, fill with average for all boroughs

final_df['Are gangs'] = final_df['Are gangs'].fillna(final_df['Are gangs'].mean())
final_df['Are you wo'] = final_df['Are you wo'].fillna(final_df['Are you wo'].mean())
final_df['How safe d'] = final_df['How safe d'].fillna(final_df['How safe d'].mean())
final_df['Are you sa'] = final_df['Are you sa'].fillna(final_df['Are you sa'].mean())
```

```python
# import centrality data, adjust to match length of final dataframe, set CRS, add to final dataframe

trafalgar_square = gpd.read_file('../../../big_files/trafalgar_square_gdf_2.shp')
trafalgar_square = trafalgar_square.drop(columns = ['FID'])
trafalgar_square_1 = trafalgar_square.loc[:len(final_df) -1]
trafalgar_square_1 = trafalgar_square_1.set_crs(epsg = 27700)
trafalgar_square_1 = trafalgar_square_1.to_crs(epsg = 27700)
final_df['dist_traf_square'] = final_df.distance(trafalgar_square_1)
```

```python
# import crime data grouped by LSOA, merge with final dataframe, fill those few LSOAs where there was no recorded crime

grouped_crime = pd.read_csv('../../../big_files/grouped_crime.csv')
grouped_crime = grouped_crime.drop(columns = ['Unnamed: 0'])
final_df = pd.merge(final_df, grouped_crime, how = 'left', left_on = 'lsoa', right_on = 'lsoa')
final_df['crime_count_lsoa'] = final_df['crime_count_lsoa'].fillna(0)
final_df = final_df.drop_duplicates(subset = ['postcode'])
```

```python
# import deprivation data and merge with final dataframe

deprivation = pd.read_csv('../../../big_files/deprivation_clean.csv')
deprivation = deprivation.drop(columns = ['Unnamed: 0'])
final_df = pd.merge(final_df, deprivation, how = 'left', left_on = 'lsoa', right_on = 'lsoa')
```

```python
# import gun crime data and merge with final dataframe, fill those wards that didn't record any gun crime

gun = pd.read_csv('../../../big_files/gun.csv')
gun = gun.loc[:, ['ward_code', 'total_gun_crimes']]
final_df = pd.merge(final_df, gun, how = 'left', left_on = 'ward_code', right_on = 'ward_code')
final_df['total_gun_crimes'] = final_df['total_gun_crimes'].fillna(0)
```

```python
# import knife crime data and merge with final dataframe, fill missing values (City of London)

knife = pd.read_csv('../../../big_files/knife.csv')
knife = knife.loc[:, ['ward_code', 'total_knife_crimes']]
final_df['total_knife_crimes'] = final_df['total_knife_crimes'].fillna(final_df['total_knife_crimes'].mean())
final_df = pd.merge(final_df, knife, how = 'left', left_on = 'ward_code', right_on = 'ward_code')
```

```python
# merge ward and lsoa data, ready to aggregate data on estate agents, delicatessens etc

lsoa = lsoa.rename(columns = {'LSOA11CD': 'lsoa'})
ward_lsoa = pd.merge(ward_lsoa, lsoa, how = 'left', left_on = 'lsoa', right_on = 'lsoa')
ward_lsoa = gpd.GeoDataFrame(ward_lsoa, geometry = ward_lsoa['geometry'])
ward_lsoa = ward_lsoa.set_crs(epsg = 27700)
ward_lsoa = ward_lsoa.to_crs(epsg = 27700)
```

```python
# import estate agent data, set CRS, merge with ward/lsoa data, groupby wards to find total/ward, merge with final dataframe
# some wards don't have any estate agents

estate_agents = gpd.read_file('../../../big_files/estate_agents_gdf.shp')
estate_agents = estate_agents.set_crs(epsg = 4326)
estate_agents = estate_agents.to_crs(epsg = 27700)
estate_agents_ward = gpd.sjoin(estate_agents, ward_lsoa, how = 'left', op = 'intersects')
ea_wards = estate_agents_ward.groupby('ward_code')['postcode'].count().reset_index()
ea_wards = ea_wards.rename(columns = {'postcode': 'ea_count'})
final_df = pd.merge(final_df, ea_wards, how = 'left', left_on = 'ward_code', right_on = 'ward_code')
final_df['ea_count'] = final_df['ea_count'].fillna(0)
```

```python
# import delicatessen data, merge with ward/lsoa data, groupby wards to find total/ward, merge with final df
# some wards don't have any delicatessens

delicatessens_gdf = gpd.read_file('../../../big_files/delicatessens_gdf.shp')
delicatessens_gdf = delicatessens_gdf.drop(columns = ['name', 'Unnamed_ 0', 'latitude', 'longitude'])
delicatessen_wards = ward_lsoa.drop(columns = ['lsoa', 'ward_name', 'borough'])
delicatessen_wards = delicatessen_wards.drop_duplicates()
delicatessen_wards = gpd.sjoin(delicatessen_wards, delicatessens_gdf, how = 'left', op = 'intersects')
delicatessen_wards = delicatessen_wards.drop_duplicates(subset = ['ward_code'])
grouped_delis = delicatessen_wards.groupby('ward_code')['postcode'].count().reset_index()
grouped_delis = grouped_delis.rename(columns = {'postcode': 'deli_count'})
final_df = pd.merge(final_df, grouped_delis, how = 'left', left_on = 'ward_code', right_on = 'ward_code')
```

```python
# import florist data, merge with ward/lsoa data, groupby wards to find total/ward, merge with final df
# some wards don't have any florists

florists = gpd.read_file('../../../big_files/florists.shp')
florists_wards = ward_lsoa.drop(columns = ['lsoa', 'ward_name', 'borough'])
florists_wards = gpd.sjoin(florists_wards, florists, how = 'left', op = 'intersects')
florists_wards = florists_wards.drop_duplicates(subset = ['ward_code'])
grouped_florists = florists_wards.groupby('ward_code')['postcode'].count().reset_index()
grouped_florists = grouped_florists.rename(columns = {'postcode': 'florist_count'})
final_df = pd.merge(final_df, grouped_florists, how = 'left', left_on = 'ward_code', right_on = 'ward_code')
```

```python
# import restaurant data, merge with ward/lsoa data, groupby wards to find total/ward, merge with final df
# some wards don't have any restaurants

restaurants_gdf = gpd.read_file('../../../big_files/restaurants_gdf.shp')
restaurants_wards = ward_lsoa.drop(columns = ['lsoa', 'ward_name', 'borough'])
restaurants_wards = restaurants_wards.drop_duplicates(subset = ['ward_code'])
restaurants_wards = gpd.sjoin(restaurants_wards, restaurants_gdf, how = 'left', op = 'intersects')
grouped_restaurants = restaurants_wards.groupby('ward_code')['postcode'].count().reset_index()
grouped_restaurants = grouped_restaurants.rename(columns = {'postcode': 'restaurant_count'})
final_df = pd.merge(final_df, grouped_restaurants, how = 'left', left_on = 'ward_code', right_on = 'ward_code')
```

```python
# import location data for 'outstanding' schools, including 500m radius around them
# separate main dataframe into properties that fall inside these areas, and those that don't

schools_500m = gpd.read_file('../../../big_files/schools_500m.shp')
inside_500m = gpd.sjoin(final_df, schools_500m, how = 'inner', op = 'intersects')
inside_500m = inside_500m.drop_duplicates(subset = ['postcode'])
ind_500 = list(inside_500m.index)
outside_500m = final_df.drop(ind_500)
outside_500m['school_pos'] = 'None'
final_df = pd.concat([inside_500m, outside_500m])
final_df = final_df.drop(columns = ['index_right'])
```

```python
# import borough shape files ready for visualising property 'move'

boroughs = gpd.read_file('../../../big_files/London_Borough_Excluding_MHW.dbf')
```

```python
# a function to 'move' a property from one postcode to another, calculate the change in value and visualise the shift


def convert_postcode(first_postcode, second_postcode, property_type):
    
    """
    'Moves' a property from one postcode to another, calculating the change in value and visualizing the move.
    
    Takes three positional arguments:
    
    first_postcode (str), the postcode the property is moving from
    second_postcode (str), the postcode the property is moving to
    property_type (str), the type of the property
    
    returns matplotlib fig, ax objects
    
    
    """
    
    if property_type.lower() == 'flat' or property_type.lower() == 'apartment':
        df = all_postcodes_df[all_postcodes_df['prop_type'] == 'F']
    elif property_type.lower() == 'semi' or property_type.lower() == 'semi-detached' or property_type.lower() == 'semi detached':
        df = all_postcodes_df[all_postcodes_df['prop_type'] == 'S']
    elif property_type.lower() == 'terrace' or property_type.lower() == 'terraced':
        df = all_postcodes_df[all_postcodes_df['prop_type'] == 'T']
    elif property_type.lower() == 'detached':
        df = all_postcodes_df[all_postcodes_df['prop_type'] == 'D']
    else:
        return 'please enter valid property type'
    
    
    try:
        row_1 = df[df['postcode'] == first_postcode]
        price_1 = row_1['price'].values
        first_borough = row_1['borough'].values
        first_geom = row_1['geometry'].reset_index()
    except:
        return 'first postcode not valid'
    
    
    try:
        row_2 = df[df['postcode'] == second_postcode]
        price_2 = row_2['price'].values
        second_borough = row_2['borough'].values
        second_geom = row_2['geometry'].reset_index()
    except:
        return 'second postcode not valid'
    
    
    pct_change = (price_2 - price_1) * 100/ price_1
    dist = first_geom.distance(second_geom).values
    
    
    print(f'Moving the {property_type.lower()} from {first_postcode} in {first_borough[0]} to {second_postcode} in {second_borough[0]} is a journey of {int(round(dist[0], 0))} meters. \nIts value would change by {int(pct_change)}%')
    

    fig, ax = plt.subplots(figsize = (15,15))
    ax.axis('off')
    
    boroughs.plot(ax = ax, edgecolor = 'black', alpha = 0.2)
    
    
    row_1.plot(ax = ax, color = 'red')
    row_2.plot(ax = ax, color = 'red')
    
    x_values = [row_1['geometry'].x, row_2['geometry'].x]
    y_values = [row_1['geometry'].y, row_2['geometry'].y]
    
    ax.text(x= row_1['geometry'].x, y = row_1['geometry'].y + 1000, s = first_postcode, fontsize = 14, ha = "center")
    ax.text(x= row_2['geometry'].x, y = row_2['geometry'].y + 1000, s = second_postcode, fontsize = 14, ha = "center")
    
    plt.plot(x_values, y_values, linewidth = 1, linestyle = "-", color = "red")
    plt.show()
    
    
    return fig, ax
```
