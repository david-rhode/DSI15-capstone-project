**18 VISUALISE PERFORMANCE OF MOST SUCCESSFUL MODEL**

- The most successful model was a Random Forest Regressor, with an r2 score of slightly over 0.52.

- It would be helpful now to have a look at its Feature Importances to see where it chooses to split, and also its residual errors.


```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
import joblib
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
# import most successful Random Forest model and transaction dataframe

rf = joblib.load('../../../big_files/rf_500.jlib')
latest_df = pd.read_csv('../../../big_files/latest_df.csv')
latest_df = latest_df.drop(columns = ['Unnamed: 0'])
```

```python
# prepare data and get Random Forest model to make predictions on X_test

updated_features = ['prop_type', 'fsm_lsoa', 'ea_in_ward', 'avg_airbnb', 'airbnb_tot', 'crime_lsoa', 'deli_count', 'flor_count', 'rest_count', 'income_rank_pos','employment_rank_pos', 'education_rank_pos', 'health_dep_score', 'crime_rank_pos', 'housing_rank_pos', 'living_env_pos', 'zone_50m','zone_200m', 'zone_400m', 'zone_800m', 'zone_1600m', 'zone_2400m', 'dist_traf', 'gang_prob_pos', 'crime_worry_pos', 'safety_fears_pos', 'satisfaction_pos', 'gun_crime', 'knife_crime', 'good_school', 'tube_zone', 'any_tube']

updated_df = latest_df[updated_features]
updated_df['price'] = latest_df['price']


y = updated_df.pop('price')
X = updated_df

# encode categorical columns

X = pd.get_dummies(X, drop_first=True)

#train test split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1)

# standardize the data

scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns)

test_predictions = rf.predict(X_test)
```

```python
# check which features the model is splitting on

fig, ax = plt.subplots(figsize = (15, 15))
rf_df = pd.DataFrame(list(zip(X_train.columns, rf.feature_importances_))
            ).sort_values(by=1, ascending=False
                         ).plot(kind='barh', x=0, y=1, ax = ax, legend = None)
plt.title('Feature Importances for Random Forest model')
plt.ylabel('Feature')
plt.show()
```

```python
# visualise actual prices vs predictions made by Random Forest model
# there is one big outlier which is making the graph difficult to view

fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(x = test_predictions, y = y_test)
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], lw = 2, c = 'r')
ax.text(x = 15880181, y = 147000000, s = 'SW1X 7LJ \n One Hyde Park', fontsize = 14, ha = "center")

plt.ticklabel_format(style='plain')
plt.xlabel('Sale Prices Predicted by Random Forest', fontsize = 15)
plt.ylabel('Actual Sale Prices', fontsize = 15)
plt.title('Actual Sale Prices vs Random Forest Predictions', fontsize = 17)

plt.show()
```

```python
# create clean predictions, with the outlier removed

clean_preds = pd.Series(test_predictions)
clean_test = pd.Series(y_test)
clean_test = clean_test.reset_index(drop = True)
clean_preds = clean_preds.reset_index(drop = True)
clean_test = clean_test.drop(1955)
clean_preds = clean_preds.drop(1955)
```

```python
# remove outlier and replot actuals vs predictions
# visualise actual prices vs predictions made by Random Forest model


fig, ax = plt.subplots(figsize=(10, 10))

ax.scatter(x = clean_preds, y = clean_test)

ax.plot([clean_test.min(), clean_test.max()], [clean_test.min(), clean_test.max()], lw = 2, c = 'r')
plt.ticklabel_format(style='plain')
plt.xlabel('Sale Prices Predicted by Random Forest', fontsize = 15)
plt.ylabel('Actual Sale Prices', fontsize = 15)
plt.title('Outlier Removed, Actual Sale Prices vs Random Forest Predictions', fontsize = 17)

plt.show()
```

```python
# plot the errors vs the sale price

epsilon = clean_preds - clean_test
epsilon_df = pd.DataFrame({'errors': epsilon, 'price': clean_test})
filt = epsilon > 0
epsilon_df.loc[filt, 'colour'] = 'yellow'
epsilon_df.loc[~filt, 'colour'] = 'black'



fig, ax = plt.subplots(figsize=(10, 10))
sns.regplot(x = 'errors', y = 'price', data = epsilon_df, fit_reg=False, scatter_kws = {'facecolors': epsilon_df['colour']})
plt.ticklabel_format(style='plain')
plt.ylabel('Actual Sale Price', fontsize = 15)
plt.axvline(linewidth = 2, color = 'r')
plt.xlabel('Error produced by Random Forest', fontsize = 15)
plt.title('Sale Price vs Error for Random Forest', fontsize = 17);
```

**CONCLUSION**

- The model tends to overvalue the cheaper properties, and undervalue the more expensive ones. 

- At the highest price points, it can undervalue them severely.

- This is likely due to interactions between the geographic features engineered from postcodes, and other missing data - most likely the overall quality of properties.

- The features work well in their own right, but to get the most value from them they should be layered on top of more traditional data (size, overall quality, num bedrooms etc)
